```python
csv_df = pd.read_csv('weather_data.csv', skiprows=1) # skip top rows
csv_df
```

```python
# cleaning of data (can set NaN with unusefull or blank etc data)
csv_df = pd.read_csv('weather_data.csv', na_values=['rain', 26, '10/1/2022', 5])
csv_df
```

```python
# cleaning of data (can set NaN with unusefull or blank etc data) 
# na_values can be dictionary for specific column data change
csv_df = pd.read_csv('weather_data.csv', na_values={
    'temperature': ['rain', 26, '10/1/2022'],
    'windspeed': ['rain', 5]
})
csv_df
```

# Read Excel

```python
ex_df = pd.read_excel('weather_dataset.xlsx')
# ex_df = pd.read_excel('weather_dataset.xlsx', 'Sheer1') # read csv with specific sheet form multiple sheet
ex_df
```

```python
# Define a user define function for converting column data 
def convert_event_cell(cell):
    if cell == 'snow':
        return 'Its Snow'
    return cell
def convert_temperature_cell(cell):
    if cell == 26:
        return 100
    return cell


# user define function use for converting excel cell with converters properties
ex_df = pd.read_excel('weather_dataset.xlsx', 'Sheet1', converters={
    'event': convert_event_cell,
    'temperature': convert_temperature_cell
})
ex_df
# Note: this is not good way, its only demo practis
```

# Write excel

```python
# writing data in excel with sheet_name
ex_df.to_excel('new.xlsx', sheet_name='weather')
```

```python
# Can set starting cell in excel for writing
ex_df.to_excel('new.xlsx', sheet_name='weather', startcol=1, startrow=2)
```

```python
# write excel without index
ex_df.to_excel('new.xlsx', sheet_name='weather', index=False)
```

# Multisheet write excel

```python
# multi sheet create form same or different dataframe 

with pd.ExcelWriter('multisheet.xlsx') as writer:
    ex_df.to_excel(writer, sheet_name='excel_weather')
    csv_df.to_excel(writer, sheet_name='csv_weather')
```
