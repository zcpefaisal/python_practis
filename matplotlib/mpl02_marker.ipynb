```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoint = np.array([3,8,1,10])
```

```python
plt.plot(ypoint, marker='o') # marker show as circle
plt.show()
```

```python
# You can choose any of these markers:
# 'o' -> Circle
# '*' -> Star
# '.' -> Point
# ',' -> Pixel
# 'x' -> X
# 'X' -> X (filled)
# '+' -> Plus
# 'P' -> Plus (filled)
# 's' -> Square
# 'D' -> Diamond
# 'd' -> Diamond (thin)
# 'p' -> Pentagon
# 'H' -> Hexagon
# 'h' -> Hexagon
# 'v' -> Triangle Down
# '^' -> Triangle Up
# '<' -> Triangle Left
# '>' -> Triangle Right
# '1' -> Tri Down
# '2' -> Tri Up
# '3' -> Tri Left
# '4' -> Tri Right
# '|' -> Vline
# '_' -> Hline
plt.plot(ypoint, marker='*') # marker show as start
plt.show()
```

```python
# Format Strings fmt ( marker|line|color )

# Line Reference
# '-' -> Solid line
# ':' -> Dotted line
# '--' -> Dashed line
# '-.' -> Dashed/dotted line

# Color reference
# 'r' -> Red
# 'g' -> Green
# 'b' -> Blue
# 'c' -> Cyan
# 'm' -> Magenta
# 'y' -> Yellow
# 'k' -> Black
# 'w' -> White

plt.plot(ypoint, 'o:r') # here o=circle marker, :=line style dotted, r=red color
plt.show()
```

```python
# Marker Size

# ms = markersize
# mec = markeredgecolor
# mfc = markerfacecolor

plt.plot(ypoint, marker = 'o', ms = 20)
plt.show()
```

```python
plt.plot(ypoint, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
plt.plot(ypoint, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
plt.plot(ypoint, marker = 'o', ms = 20, mfc = 'r', mec = 'r')
plt.show()
```

```python
plt.plot(ypoint, marker = 'o', ms = 20, mfc = '#4CAF50', mec = '#ccc')
plt.show()
```
