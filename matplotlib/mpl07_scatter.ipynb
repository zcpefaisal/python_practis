```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
```

```python
# The scatter() function plots one dot for each observation. 
# It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:
plt.scatter(x,y)
plt.show()
```

```python
# Can set own color for each scatter plot with the color or the c argument
# Data set 1
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y, color='red')

# Data set 2
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x2, y2, color='green')

plt.show()
```

```python
# Color Each Dot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y, c=colors) #Note: Here cannot use the "color" argument for this, only can use "c" argument.
plt.show()
```

```python
# Use the ColorMap
# can specify the colormap with the keyword argument "cmap" with the value of the colormap, 
# in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y, c=colors, cmap='viridis')

plt.colorbar() # can include the colormap in the drawing by including the plt.colorbar()
plt.show()
```
