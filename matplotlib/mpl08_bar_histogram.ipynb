```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([10,4,9,2])
```

```python
plt.bar(x,y) # Vartical Bar
plt.show()
```

```python
plt.barh(x,y) # Horizontal Bar
plt.show()
```

```python
plt.bar(x,y, color='green') # Vartical Bar
plt.show()
```

```python
plt.bar(x,y, color='#ccc', width=0.25) # Vartical Bar
plt.show()
```

```python
plt.barh(x,y, color='#000', height=0.25) # Vartical Bar
plt.show()
```
