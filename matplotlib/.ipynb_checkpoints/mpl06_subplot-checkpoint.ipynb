```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The subplot() function takes three arguments that describes the layout of the figure.
# The layout is organized in rows and columns, which are represented by the first and second argument.
# The third argument represents the index of the current plot.

x = np.array([0,1,2,3,4])
y = np.array([3,8,1,10,1])

x1 = np.array([0,1,2,3])
y1 = np.array([10,20,30,40])
```

```python
#this plot has one row and two column for subplot
# subplot third param represent in which position (index) will show (here 2=second position, 1=first position)
plt.subplot(1,2,2)
plt.plot(y)
plt.title("Second position")
plt.grid()


plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title("First position")

plt.show()
```

```python
#this plot has two row and one column for subplot
# subplot third param represent in which position (index) will show (here 2=second position, 1=first position)
plt.subplot(2,1,2)
plt.plot(x,y)
plt.ylabel("Second position")
plt.grid()


plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.ylabel("First position")

plt.show()
```

```python
#this plot has two row and one column for subplot
# subplot third param represent in which position (index) will show (here 2=second position, 1=first position)
plt.subplot(2,3,1)
plt.plot(x,y)
plt.title("Posotion(1): 1:1")

plt.subplot(2,3,2)
plt.plot(x1,y1)
plt.title("Posotion(2): 1:2")

plt.subplot(2,3,3)
plt.plot(x1,y1)
plt.title("Posotion(3): 1:3")


plt.subplot(2,3,4)
plt.plot(x1,y1)
plt.xlabel("Posotion(4): 2:1")

plt.subplot(2,3,5)
plt.plot(x,y)
plt.xlabel("Posotion(5): 2:2")

plt.subplot(2,3,6)
plt.plot(x1,y1)
plt.xlabel("Posotion(6): 23")

supFontStyle = {'color':'green'}
plt.suptitle("This is Super Title", fontdict=supFontStyle)

plt.show()
```
