```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
```

```python
plt.title("Sports Watch Data")
plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x,y)
plt.grid() # can use the grid() function to add grid lines to the plot.
plt.show()
```

```python
plt.plot(x,y)
# can use the axis parameter in the grid() function to specify which grid lines to display.
# values are: 'x', 'y', and 'both'. Default value is 'both'.
plt.grid(axis='x') 

plt.show()
```

```python
plt.plot(x,y)
# can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).
plt.grid(color="green", linestyle="--", linewidth=1.5) 

plt.show()
```
