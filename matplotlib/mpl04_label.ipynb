```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
```

```python
plt.plot(x,y)
plt.title("Sports Watch Data") # title text show anything
plt.xlabel("Avarage pulse") # x-asix wise text show anything
plt.ylabel("Calorie burnage") # y-asix wise text show anything
plt.show()
```

```python
fontStyle1 = {'family':'serif', 'color':'blue', 'size':15} 
fontStyle2 = {'family':'serif', 'color':'green', 'size':15} 
fontStyle3 = {'family':'serif', 'color':'red', 'size':15} 
# can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties
# can use the "loc" parameter in title() to position the title, values are: 'left', 'right', and 'center'. Default is 'center'.
plt.title("Sports Watch Data", fontdict = fontStyle1, loc='left') # title text show anything
plt.xlabel("Avarage pulse", fontdict = fontStyle2) # x-asix wise text show anything
plt.ylabel("Calorie burnage", fontdict = fontStyle3) # y-asix wise text show anything
plt.plot(x,y)
plt.show()
```
