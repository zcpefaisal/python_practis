```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([2,5,4,8,3])
```

```python
plt.plot(ypoints, linestyle='dotted')
plt.show()
```

```python
# Shorter Syntax
# linestyle = ls
# dotted = :
# dashed = --

plt.plot(ypoints, ls = '--') 
plt.show()
```

```python
# Line Styles

# 'solid' (default) -> '-'
# 'dotted' -> ':'
# 'dashed' -> '--'
# 'dashdot' -> '-.'
# 'None' -> '' or ' '
plt.plot(ypoints)
plt.show()
```

```python
# Line Color
# color = c
plt.plot(ypoints, color='g')
plt.show()
```

```python
# Line Width
# linewidth = lw
plt.plot(ypoints, lw='10')
plt.show()
```

```python
# Multiple Lines

ypoints1 = np.array([1,8,3,10,2,5])
ypoints2 = np.array([9,2,5,0,10,3])
plt.plot(ypoints1)
plt.plot(ypoints2)
plt.show()
```

```python
# x-axis & y-asix wise myltiple line in same plot
x1 = np.array([0, 5, 10, 15])
y1 = np.array([3, 8, 1, 10])

x2 = np.array([1, 4, 7, 10])
y2 = np.array([6, 2, 8, 3])

plt.plot(x1, y1, x2, y2)
plt.show()
```
