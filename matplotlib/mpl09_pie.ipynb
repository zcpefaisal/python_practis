```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([80,40,20,10])
```

```python
# pie() start form x-axis, and work as anticlock wise (counterclockwise)
# The size of each wedge is determined by comparing the value with all the other values, by using this formula:
# The value divided by the sum of all values: x/sum(x)

# print(sum(y)) 150
plt.pie(y) 
plt.show()
```

```python
myLabels = ['Apple', 'Banana', 'Mango', 'cherry']
plt.pie(y, labels=myLabels)
plt.show()
```

```python
# default start angle is at the x-axis, but you can change the start angle by specifying a "startangle" parameter.
plt.pie(y, labels=myLabels, startangle=180)
plt.show()
```

```python
# "explode" parameter, if specified, and not None, must be an array with one value for each wedge.
# Each value represents how far from the center each wedge is displayed
myExplode = np.array([0,0.1,0,0])
plt.pie(y, explode=myExplode, shadow=True) # shadow to the pie chart by setting the shadows parameter to True
plt.show()
```

```python
# can set the color of each wedge with the "colors" parameter.
myColor = ['red', 'black', 'green', 'cyan']
plt.pie(y, labels=myLabels, colors=myColor)
plt.show()
```
